```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
import pandas as pd
```

```python
data = pd.read_csv('K_ Means.csv')
data
```

```python
data.head()
```

```python
# Assuming the last column contains labels
X = data.iloc[:, :-1]  # Features
y_true = data.iloc[:, -1]  # True labels
```

```python
# Create a KMeans instance
kmeans = KMeans(n_clusters=3)  # Change the number of clusters as needed
```

```python
# Fit the KMeans model to your data
kmeans.fit(X)
```

```python
# Obtain the predicted labels
y_pred = kmeans.labels_
```

```python
# Calculate accuracy
accuracy = metrics.accuracy_score(y_true, y_pred)
```

```python
precision = metrics.precision_score(y_true, y_pred, average='weighted', zero_division=0)
```

```python
# Calculate precision
precision = metrics.precision_score(y_true, y_pred, average='weighted')
```

```python
# Visualize the clusters
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y_pred, cmap='viridis', edgecolor='k')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, label='Centroids')
plt.title('KMeans Clustering')
plt.legend()
plt.show()
```

```python
# Print accuracy and precision
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
```
